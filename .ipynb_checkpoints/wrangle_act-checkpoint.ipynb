## Gather

```python
# get WeRateDogs twitter archive downloaded manually

archive = pd.read_csv('twitter-archive-enhanced.csv')
```

```python
# download image-predictions file programmatically 

image_pred_link = 'https://d17h27t6h515a5.cloudfront.net/topher/2017/August/599fd2ad_image-predictions/image-predictions.tsv'
image_predictions = requests.get(image_pred_link).content
# write to image-predictions.csv if not already done
file_name = 'image-predictions.tsv'
if not os.path.exists(file_name):
    with open(file_name, 'wb') as file:
        file.write(image_predictions)

images = pd.read_csv('image-predictions.tsv', sep='\t', encoding = 'utf-8')
```

```python
# CONSUMER_KEY = 'Consumer Key (API key)'
# CONSUMER_SECRET = 'Consumer Secret (API Secret)'
# OAUTH_TOKEN = 'Access Token'
# OAUTH_TOKEN_SECRET = 'Access Token Secret'

consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

# Construct the API instance
api = tweepy.API(auth, 
                 parser = tweepy.parsers.JSONParser(), # Parse the result to Json Object
                 wait_on_rate_limit = True, # Automatically wait for rate limits to replenish
                 wait_on_rate_limit_notify = True) # Print a notification when Tweepy is waiting for rate limits to replenish
```

```python
df_list = []
error_list = []
tweets_json_dict = {}

for tweet_id in archive['tweet_id']:
    try:
        response = api.get_status(tweet_id)
        
        # Indicates approximately how many times this Tweet has been liked by Twitter users
        favorite_count = response['favorite_count']
        
        # Indicates whether this Tweet has been liked by the authenticating user
        favorited = response['favorited']
        
        # Number of times this Tweet has been retweeted
        retweet_count = response['retweet_count']
        
        # Indicates whether this Tweet has been Retweeted by the authenticating user
        retweeted = response['retweeted']
        
        # The number of followers this account currently has
        user_followers = response['user']['followers_count']
        
        # The number of public lists that this user is a member of
        user_listed = response['user']['listed_count']
        
        # user id for the tweet
        user_id = response['user']['id']
        
        temp = {'favorite_count' : favorite_count,
                       'favorited': favorited,
                       'retweet_count': retweet_count,
                       'retweeted' : retweeted,
                       'user_followers':user_followers,
                       'user_listed':user_listed,
                       'user_id':user_id,
                       'tweet_id':tweet_id}
        
        df_list.append(temp)
        
        tweets_json_dict[str(tweet_id)] = temp
        
    except Exception as e:
        print (str(tweet_id) + '_' + str(e))
        error_list.append(tweet_id)
        
```

```python
# Change type to save to json
for val in tweets_json_dict.values():
    
    for key in val:
        val[key] = str(val[key])
```

```python
if not os.path.exists('tweet_json.txt'):
    with open('tweet_json.txt','w') as outfile:
        json.dump(tweets_json_dict, outfile, ensure_ascii=False, indent=4)
```

```python
tweets_json = pd.read_json('tweet_json.txt').transpose()
```

```python
tweets_json.set_index(np.arange(0, len(tweets_json)), inplace=True)
```

#### Define

Perform string operations to extract url from `a` tag

#### Code

```python
def extract_url_from_source (s):
    # steps to move forward for the actual url
    start = s.find('href') + len('href="')
    
    # find second `"`, for the end step and add all steps untill first `"` 
    # to use it with the original string
    end = s[s.find('"')+1:].find('"') +s.find('"')+1
    
    return s[start : end]

# assign actual urls to source
archive_copy.source = archive_copy.source.apply(extract_url_from_source) 
```

```python
assert sum(archive_copy.source.str.startswith('<a'))==0
```

#### Define

Correct rating_numerator by correctly extracting them from text column.

#### Code

```python
# regex pattern to extract numerator and denominator of ratings
pat = r'[-+]?\d+\.\d+/\d+|\d+\d+/\d+'
correct_rating = []
incorrect_ids = ['883482846933004288', '681340665377193984']

# extract correct ratings for incorrect ratings
for ids in incorrect_ids:
    txt = archive_copy[archive_copy.tweet_id == ids].text.tolist()[0]
    correct_rating.append((ids, re.findall(pat, txt)[0]))
    
# assign correct ratings 
for vals in correct_rating:
    ids = vals[0]
    rating = vals[1].split('/')
    num = rating[0]
    den = rating[1]
    archive_copy.loc[archive_copy.tweet_id == ids, 'rating_numerator'] = num
    archive_copy.loc[archive_copy.tweet_id == ids, 'rating_denominator'] = den
```

#### Define

Collapse four stage columns doggo, floofer, pupper, puppo into a 'stage' column with corresponding values and drop these four columns

#### Code

```python
value_vars = ['doggo', 'floofer', 'pupper', 'puppo']
id_vars = [col for col in archive_copy.columns if col not in value_vars]
value_name = 'stage' 
```

```python
archive_copy['stage'] = archive_copy[value_vars].apply(lambda x : ''.join(x), axis=1)
```

```python
keys = [val for val in archive_copy.stage.unique()]
to_replace = {}
doggo = 'doggo'
floofer = 'floofer'
pupper = 'pupper'
puppo = 'puppo'
for key in keys:
    if doggo in key:
        to_replace[key]=doggo
    elif floofer in key:
        to_replace[key] = floofer
    elif pupper in key:
        to_replace[key]=pupper
    elif puppo in key:
        to_replace[key]=puppo
    else:
        to_replace[key]='None'
to_replace
```

```python
archive_copy.stage.replace(to_replace=to_replace, inplace=True)
```

```python
archive_copy.drop(value_vars, axis=1, inplace=True)
```

### Ratings over Time

#### Analysis
